```python
import src
import pandas as pd
import matplotlib.pyplot as plt

Dataset, CycleOfLife = src.get_data()
labeled_df = src.get_quantile(CycleOfLife)
features = src.extract_battery_features(Dataset)

features1 = features.drop(axis='columns', columns="cell_id", )

features1.corr()
```

```python

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

x_name = "peak_height_dV_dQ"
y_name = "time_above_3_5V_cycle1"

x = features1[x_name].values.reshape(-1, 1)
y = features1[y_name].values

model = LinearRegression()
model.fit(x, y)

y_pred = model.predict(x)
r2 = r2_score(y, y_pred)

plt.scatter(
    x=features1[x_name],
    y=features1[y_name],
    label="Data"
)

plt.plot(
    features1[x_name],
    y_pred,
    color="red",
    label=f"Linear fit, $R^2$ = {r2:.3f}"
)

plt.title(f'Correlation {x_name} and {y_name}')
plt.xlabel(f"{x_name}")
plt.ylabel(f"{y_name}")
plt.legend()
plt.show()
```
